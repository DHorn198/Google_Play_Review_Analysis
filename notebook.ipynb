```python
import pandas as pd
```

```python
data = pd.read_csv('googleplaystore_user_reviews.csv') #import file needed   
```

```python
print(data.head()) #Print the first 5 rows
```

```python
print(data.isnull().sum()) #Checking to see if there are null values
```

```python
data = data.dropna() #dropping null values
```

```python
print(data.isnull().sum()) #checking to make sure all null values are gone
```

```python
import nltk
```

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sentiments = SentimentIntensityAnalyzer()
```

```python
data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data ["Translated_Review"]]
data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data ["Translated_Review"]]
data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data ["Translated_Review"]]
```

```python
print(data.head())
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15, 10))
```

```python
sns.scatterplot(x = data['Sentiment_Polarity'], y = data['Sentiment_Subjectivity'], hue = data['Sentiment'], edgecolor = 'white', palette = 'twilight_shifted_r')  #Creating Scatter Plot
plt.title("Google Play Store Reviews Sentiment Analysis", fontsize = 20) #Creating title 
plt.show() #Showing scatter plot with title
```
